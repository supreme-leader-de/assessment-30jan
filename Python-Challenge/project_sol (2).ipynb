```python
import pandas as pd

df = pd.read_csv("sales_data.csv",engine="python",header=None,names=["Order ID","Product","Quantity Ordered","Price Each","Order Date","Street","City","State Zip"])
```

```python
df.columns
```

```python
df
```

```python
df = df.iloc[1:].reset_index(drop=True)
```

```python
df
```

```python
df["Purchase Address"] = (df["Street"].astype(str) + ", " + df["City"].astype(str) + ", " +df["State Zip"].astype(str))
```

```python
df=df.drop(columns=["Street", "City", "State Zip"])
```

```python
df
```

```python
df = df.dropna(axis=0, subset=['Order ID'])
df = df.dropna(axis=0, subset=['Product'])
```

```python
df["Price Each"] =df["Price Each"].astype(float)
```

```python
mean_p = df.groupby("Product")["Price Each"].mean()
df["Price Each"] = df["Price Each"].fillna(df["Product"].map(mean_p))
```

```python
df
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df
```

```python
#1
df["Total Sales"]=(df["Quantity Ordered"].astype(float))*df["Price Each"]
```

```python
df
```

```python
#2
df["month"] = df["Order Date"].dt.month
monthly_rev= df.groupby("month")["Total Sales"].sum()
```
